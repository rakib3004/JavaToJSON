```python
import os
import json
import openai
import Levenshtein
from openai import OpenAI
import math


client = OpenAI(
api_key = 'sk-open-api-key',
)

def levenshtein_distance(filename, gpt_completed_json, original_complete_json):
    print(gpt_completed_json)
    levenshtein_distance = Levenshtein.distance(
    str(gpt_completed_json), str(original_complete_json)
    )

    print(filename, levenshtein_distance)



def complete_json_from_gpt(filename, original_complete_json, incomplete_json):
    prompt = "There are some method's calls are missing, fill up them and complete the JSON file"

    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": f"{incomplete_json} {prompt}",
            }
        ],
        model="gpt-3.5-turbo",
    )
    gpt_completed_json = chat_completion.choices[0].message.content
    levenshtein_distance(filename, gpt_completed_json, original_complete_json)


def hiding_methods(temp_json):
  incomplete_json = temp_json
  for entry in incomplete_json.get("classes", []) + incomplete_json.get("abstract_classes", []) + incomplete_json.get("interfaces", []):
        methods =  entry.get("methods", [])
        num_methods = len(methods)

        indices_to_hide = [i for i in range(0, num_methods) if i % 3 != 0]

        for index in indices_to_hide:
            methods[index]["name"] = ""


  # print(json.dumps(incomplete_json, indent=2))
  return incomplete_json

def hiding_calls(temp_json):
  incomplete_json = temp_json
  for entry in incomplete_json.get("classes", []) + incomplete_json.get("abstract_classes", []):
        methods =  entry.get("methods", [])
        num_methods = len(methods)

        indices_to_hide = [i for i in range(0, num_methods) if i % 3 != 0]

        for index in indices_to_hide:
            methods[index]["calls"] = []



  # print(json.dumps(incomplete_json, indent=2))
  return incomplete_json


def hiding_relationships(temp_json):
  total_relationships = 0

  incomplete_json = temp_json
  for entry in incomplete_json.get("classes", []) + incomplete_json.get("abstract_classes", []):
        relationships = entry.get("relationships", [])
        total_relationships += len(relationships)

  relationship_index=0
  for entry in incomplete_json.get("classes", []) + incomplete_json.get("abstract_classes", []):
        relationships = entry.get("relationships", [])

        if(relationship_index%3!=0):
          relationships[0]["relationships"] = []
          relationship_index+=1



  # print(json.dumps(incomplete_json, indent=2))
  return incomplete_json



def hiding_attributes(original_complete_json):
  temp_json = original_complete_json
  # incomplete_json = hiding_methods(temp_json)
  incomplete_json = hiding_calls(temp_json)
  # incomplete_json = hiding_relationships(temp_json)
  return incomplete_json


def read_json_files(folder_path):

    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            file_path = os.path.join(root, filename)

            if filename.endswith(".json") and os.path.isfile(file_path):
                with open(file_path, 'r') as file:
                    try:
                        original_complete_json = json.load(file)
                        incomplete_json = hiding_attributes(original_complete_json)
                        complete_json_from_gpt(filename, original_complete_json, incomplete_json)

                    except json.JSONDecodeError as e:
                        print(f"Error decoding JSON in file {file_path}: {e}")


folder_path = '/content/drive/MyDrive/SPL3/WilliamMary/JSON/SPL3'
read_json_files(folder_path)

```

# Packages Info

```python
import os
import json
import openai
import Levenshtein
from openai import OpenAI
import math
import pandas as pd
import csv


def collect_information(json_data):
    package_name = json_data["package"][0] if "package" in json_data else ""

    grand_total_methods = 0
    grand_total_calls = 0
    grand_total_relationships = 0



    for class_entry in json_data.get("classes", []) + json_data.get("abstract_classes", []) + json_data.get("interfaces", []) :
        total_methods = len(class_entry.get("methods", []))
        total_calls = sum(len(method.get("calls", [])) for method in class_entry.get("methods", []))
        total_relationships = len(class_entry.get("relationships", []))

        grand_total_methods += total_methods
        grand_total_calls += total_calls
        grand_total_relationships += total_relationships


    return package_name, grand_total_methods, grand_total_calls, grand_total_relationships






def read_json_files(folder_path):
    information_list = []
    for root, dirs, files in os.walk(folder_path):
        for filename in files:
            file_path = os.path.join(root, filename)

            if filename.endswith(".json") and os.path.isfile(file_path):
                with open(file_path, 'r') as file:
                    try:
                        json_data = json.load(file)
                        package_name, grand_total_methods, grand_total_calls, grand_total_relationships =collect_information(json_data)
                        information_list.append([package_name, grand_total_methods, grand_total_calls, grand_total_relationships])

                    except json.JSONDecodeError as e:
                        print(f"Error decoding JSON in file {file_path}: {e}")

    with open(f'/content/drive/MyDrive/SPL3/WilliamMary/Report/{repository_name}.csv', 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['Package Name', 'Total Methods', 'Total Calls', 'Total Relationships'])
        csv_writer.writerows(information_list)

repository_name = 'software-testing'
folder_path = f'/content/drive/MyDrive/SPL3/WilliamMary/JSON/{repository_name}'
read_json_files(folder_path)


```
